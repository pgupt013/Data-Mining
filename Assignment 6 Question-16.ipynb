```python
# import the libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage  
#input similarity matrix
mat = np.array([[1.0, 0.10, 0.41,0.55,0.35], 
                [0.10, 1.0,0.64,0.47,0.98], 
                [0.41,0.64,1.0, 0.44,0.85],
                [0.55,0.47,0.44, 1.0,0.76],
                [0.35,0.98,0.85, 0.76,1.0]])
```

```python
#plot the matrix points
labels = range(1, 6)  
plt.figure(figsize=(7, 5))  
plt.subplots_adjust(bottom=0.1)  
plt.scatter(mat[:,0],mat[:,1], label='True Position')
for label, x, y in zip(labels, mat[:, 0], mat[:, 1]):  
    plt.annotate(
        label,
        xy=(x, y), xytext=(-3, 3),
        textcoords='offset points', ha='left', va='bottom')
plt.show() 
```

```python
#plotting the Single link dendrogram
dist_mat = mat
linkage_matrix = linkage(dist_mat, "single")
dendrogram(linkage_matrix, color_threshold=1, labels=["1", "2","3","4","5" ],show_leaf_counts=True)
plt.title=("test")
plt.show()
```

```python
#plotting the complete link dendrogram
linkage_matrix = linkage(mat, "complete")
labelList = range(1, 6)
plt.figure(figsize=(7, 5))  
dendrogram(linkage_matrix,  
            orientation='top',
            labels=labelList,
            distance_sort='descending',
            show_leaf_counts=True)
plt.show() 
```
